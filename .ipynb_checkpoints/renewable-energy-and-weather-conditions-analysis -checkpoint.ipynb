# Generating Desired Column

```python
time = data['Time'].tolist()
```

```python
time
```

```python
year = []
times = []
for i in range(len(time)):
    year.append(time[i][0:4])
    times.append(time[i][11:])
```

```python
data['Time'] = pd.to_datetime(data['Time'])
```

```python
data['Year'] = year
data['instance'] = times
```

```python
year_int = [eval(i) for i in year]
data['Year'] = year_int
```

```python
data["Time"] = pd.to_datetime(data["Time"])
```

```python
data
```

## Finding Relations

```python
figure, ax = plt.subplots(2,1,figsize = (20,10))
plt.subplots_adjust(hspace=0.5)
sns.lineplot(x=data['Time'].dt.strftime('%Y-%m'), y='temp', data=data, ax = ax[0])
ax[0].tick_params('x', labelrotation=75)
sns.lineplot(x=data['Time'].dt.strftime('%Y-%m'), y='Energy delta[Wh]', data=data, ax = ax[1])
ax[1].tick_params('x', labelrotation=75)
figure.suptitle('Relation Between Temperature and Energy Delta over Time ', fontsize=25)
ax[0].title.set_text('Line Plot Represent the Temperature Over The Time Periiod')
ax[1].title.set_text('Line Plot Represent the Energy delta Over The Time')
plt.show()
```

```python
# This plot clearly define the relation between temperature and raise in energy consuption.
```

```python

data.set_index("Time", inplace=True)
energy_need = data["Energy delta[Wh]"].resample("D").mean()
# Energy consuption Perday Over the time
plt.plot(energy_need.index, energy_need.values)
plt.title("consuption Over Time")
plt.xlabel("Time")
plt.ylabel("Mean Energy Consuption")
plt.show()
```

```python
# This above graph indicates that there is a pattern in the consuption of energy use. 
# we need to find out every possible relation with the energy consuption so that we can move further with our analysis
plt.plot(data['hour'].unique(),data.groupby('hour').mean()['Energy delta[Wh]'])
plt.title('Mean of Energy delta[Wh] for Each Hour')
# Here ce can find the peak hour for energy uses
```

```python
plt.plot(data['month'].unique(),data.groupby('month').mean()['Energy delta[Wh]'])
plt.title('Mean of Energy delta[Wh] for Each Month')
```

```python
plt.plot(data['hour'].unique(),data.groupby('hour').mean()['temp'])
plt.title('Average Temperature for Each Hour')
# As we alredy found that there is a direct relationship between temperature and energy uses. we can also see that here
```

# CONCLUSION

1. Solar energy is the best option for this geographical area where they can fulfil their energy need as there hardly any snowfall or cloud coverage when the energy delta is in peak.
2. in thatt same duretion the GHI is also in peak.
3. It is noticed that there is a continuous decrease in rainfall so it is not viable to relay on it, as the huge capital investment is required for the construction of hydro power projects.
4. It is also noticed that the average wind speed is less than 5 m/s in this condition a wind mill is not economical further  more the wind speed is very low in those month when there is a surge in energy demand.
5. in the case of solar energy the GHI is in its maximum when the energy demand is more.
6. From all the observetion it is evident that solar energy is the best option for that region.
